```python
import matplotlib.pyplot as plt
import numpy as np

# Input: dictionary of models and their corresponding accuracy values
model_accuracies = {
    'Support Vector Classifier': 0.9785,
    'Logistic Regression': 0.9808,
    'Decision Tree': 0.9444,
    'Naive Bayes': 0.9601,
    'BERT': 0.9888  # Change these values as per your accuracy results
}

# Find the best and least best model
best_model = max(model_accuracies, key=model_accuracies.get)
worst_model = min(model_accuracies, key=model_accuracies.get)

# Plotting the results
models = list(model_accuracies.keys())
accuracies = list(model_accuracies.values())

# Create a bar plot
fig, ax = plt.subplots(figsize=(10, 6))  # Increase figure size for better clarity
bars = ax.bar(models, accuracies, color='lightblue')

# Highlight the best and worst model with different colors
for bar, model in zip(bars, models):
    if model == best_model:
        bar.set_color('blue')
    elif model == worst_model:
        bar.set_color('red')

# Add titles and labels
plt.title('Model Performance in Detecting Spam Emails', fontsize=14)
plt.xlabel('Machine Learning Models', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.ylim(0.9, 1.02)  # Accuracy scale 0.9 to a bit over 1 to create space for the numbers

# Annotating the bars with accuracy values
for i, v in enumerate(accuracies):
    plt.text(i, v + 0.005, f"{v*100:.2f}%", ha='center', va='bottom', fontsize=10)

# Add a horizontal grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
